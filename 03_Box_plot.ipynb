```python
import seaborn as sns
import pandas as pd
import numpy as np

kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived",
            y="age",
            data=kashti)
```

```python
sns.boxplot(x="survived",
            y="age", showmeans=True,
            data=kashti)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.boxplot(x="survived",
            y="age", showmeans=True,
            meanprops= {"marker": "*",
                       "markersize": "12",
                       "markeredgecolor": "red"},
            data=kashti)
plt.xlabel("How many survived", size=14),
plt.ylabel("Age (years)", size=12),
plt.title("Box Plot of Kitnay Doubey and Kitnay Bach gay", size= 12)
```
